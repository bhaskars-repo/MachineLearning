# Logistic Regression

#### Python Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
plt.style.use('seaborn')
np.random.seed(101)
```

#### Sigmoid function $\sigma(y) = 1/(1 + e^{-y})$

```python
x = np.linspace(-5, 5, 100)
y = 1/(1 + np.exp(-x))
```

```python
plt.xlim(-5.05, 5.05)
plt.ylim(-0.1, 1.1)
plt.axhline(0.5, color='lightgray', linestyle='--')
plt.axvline(0, color='lightgray', linestyle='--')
plt.annotate('$\\sigma(y) = 1/(1+e^{-y})$', (-2.5, 0.65), xytext=(-2.5, 0.65), ha='center')
plt.plot(0.0, 0.5, 'ro', markersize=10)
plt.annotate('$\\sigma(y) = 0.5$', (0.7, 0.5), xytext=(0.7, 0.5), ha='center')
plt.axhline(0.0, color='coral', linestyle='--')
plt.axhline(1.0, color='coral', linestyle='--')
plt.plot(x, y, color='limegreen')
plt.show()
```
