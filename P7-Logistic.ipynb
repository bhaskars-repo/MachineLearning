# Logistic Regression

#### Python Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use('seaborn')
np.random.seed(101)
```

#### Sigmoid function $\sigma(y) = 1/(1 + e^{-y})$

```python
x = np.linspace(-5, 5, 100)
y = 1/(1 + np.exp(-x))
```

```python
plt.xlim(-5.05, 5.05)
plt.ylim(-0.1, 1.1)
plt.axhline(0.5, color='lightgray', linestyle='--')
plt.axvline(0, color='lightgray', linestyle='--')
plt.annotate('$\\sigma(y) = 1/(1+e^{-y})$', (-2.5, 0.65), xytext=(-2.5, 0.65), ha='center')
plt.plot(0.0, 0.5, 'ro', markersize=10)
plt.annotate('$\\sigma(y) = 0.5$', (0.7, 0.5), xytext=(0.7, 0.5), ha='center')
plt.axhline(0.0, color='coral', linestyle='--')
plt.axhline(1.0, color='coral', linestyle='--')
plt.plot(x, y, color='limegreen')
plt.show()
```

#### Receiver Operating Characteristics (ROC) curve

```python
x1 = np.arange(0, 10, 0.5)
y1 = x1
x2 = np.linspace(0, 10, 50)
y2 = 9.0 * (1.0 - np.exp(-1.2*x2))
```

```python
plt.xlim(-0.5, 9.0)
plt.ylim(-0.5, 9.5)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.xticks([])
plt.yticks([])
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.plot(x1, y1, color='coral', linestyle='--')
plt.plot(x2, y2, color='limegreen')
plt.fill_between(x1, y1, color='coral', alpha=0.8)
plt.fill_between(x2, y2, color='skyblue', alpha=0.3)
plt.annotate('0', (-0.25, -0.38), xytext=(-0.25, -0.38), fontsize=14, weight='bold', ha='center')
plt.annotate('1', (8.8, -0.38), xytext=(8.8, -0.38), fontsize=14, weight='bold', ha='center')
plt.annotate('1', (-0.25, 9.0), xytext=(-0.25, 9.0), fontsize=14, weight='bold', ha='center')
plt.annotate('*', (0.25, 8.8), xytext=(0.25, 8.8), fontsize=28, weight='bold', color='red', ha='center')
plt.annotate('ROC', (1.1, 8.0), xytext=(1.1, 8.0), fontsize=15, weight='bold', ha='center')
plt.annotate('Worse', (6.5, 2.8), xytext=(6.5, 2.8), fontsize=14, ha='center')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
